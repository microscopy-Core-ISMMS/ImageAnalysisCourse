# Module 5 - Do's and Dont's

**Image Processing & Analysis: Dos and Don’ts**

• Manipulation should be done on a copy not the original data

• Simple adjustments to the entire image are usually acceptable

• Check journal’s guidances

• Ethical guidelines: http://www.ncbi.nlm.nih.gov/pubmed/20567932

• Be honest, always describe the adjustments made

• Do not hide information (brightness, contrast, gamma)

• Do not create/destroy information (interpolation, bit depth conversion)

**Understanding Noise**


- Salt and Pepper noise is caused by data loss or saturation, this type of noise contains random white (salt) and black (pepper) pixels in the image. The white pixels or spots are caused by erroneous saturation and the black pixels or spots are caused by a failed response from the sensor.

image.png

- Uniform noise can occur with low lighting, for example. This type of noise has the effect of adding or subtracting a random amount to each value when the images are captured. The range of possible values added or taken away (a) reflect the severity of the noise, and forms a probability density function (PDF) where the range is (-a to a).

# Module 7 - Filters


- Noise reduction is the process of removing as much of the noise from an image as possible - we can never remove it all perfectly.

- There are different types of noise, and so there are correspondingly different noise reduction processes that are used depending on the noise type.

- Noise reduction processes usually take the form of filters.

- These work by examining a small window of pixels called a kernel, for example 3 x 3 pixels in size, then recalculating the middle pixel based on the neighboring pixels in this window.
- This window is then moved across the image typically a pixel at a time, and applied again until the whole image has been recalculated.
- We can perform different calculations on the pixels in the kernel to handle different kinds of noise, but in all cases we are using local image statistics (such as a mean) to recalculate the center value based on the context of its neighbours.

image.png

**Common ImageJ filters**

**Smooth** (Process->Smooth)

Blurs the active image or selection. This filter replaces each pixel with the average of its 3x3 neighborhood.

                    1 1 1
                    1 1 1   *1/9
                    1 1 1
**Sharpen**

Increases contrast and accentuates detail in the image or selection, but may also accentuate noise. This filter uses the following weighting factors to replace each pixel with a weighted average of the 3x3 neighborhood.

                    -1 -1 -1
                    -1 12 -1
                    -1 -1 -1

**Find Edges**

Uses a Sobel edge detector to highlight sharp changes in intensity in the active image or selection. Two 3x3 convolution kernels (show below) are used to generate vertical and horizontal derivatives. The final image is produced by combining the two derivatives using the square root of the sum of the squares.

          1  2  1     1  0 -1
          0  0  0     2  0 -2
         -1 -2 -1     1  0 -1

         Vertical     Horizontal

# Module 9 - Colocalization

**Colocalization** consists in evaluating if two fluorophores are spatially
close

- Colocalization cannot conclude about physical interactions
(Rayleigh criterion, need to use FRET microscopy)


**Intensity-based analysis**

Scatterplot: 2D intensity histogram showing the distribution of intensity
in both channels

image.png

**Pearson’s correlation coefficient (PCC)**

PCC is the covariance of two variables divided by the product of their standard deviations. It is typically used to measure the linear correlation of the signal intensity values for two reporters.

image.png

image.png

**Costes’ automatic threshold**

Costes et al. have proposed to automatically compute a threshold in each channel:
- thresholds are initialized to the maximum intensity
- thresholds are progressively decreased
- Pearson’s correlation coefficient is computed for each value
- final thresholds minimize the contribution of noise
