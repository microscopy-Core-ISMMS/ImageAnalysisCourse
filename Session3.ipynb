**Segmentation**

Segmentation is the separation of an image into areas of interest and areas that are not of interest.

Once segmented, these objects can be further processed or analyzed for
quantitative information.

The variety of analysis: Cell counting, monitoring shape changes, quanitfying gene expressions or recording a certain activity within specific regions.

What makes biological images difficult to segment is that they are
usually very noisy, messy and of very irregular shapes. Segmentation can
be (and often is) combined with various filtering techniques to achieve
higher accuracy.


The most important principles of segmentation is dependent on two basic
properties of intensity values:
**Similarity and Discontinuity**

**Similarity-based approaches**: Thresholding and Watershed

**Discontinuity-based approaches**: Edge detection

Most fluorescence images of biological structures comprise of light object over a constant dark background. So the gray values of your image are grouped generally into two main groups – the dark, and the bright.


If you have such conveniently contrasted images in hand, one obvious way is to
extract the objects from its background by selecting a threshold, T:

            g(x,y) = 1 If f(x,y) > T
                   = 0 otherwise
where g(x,y) is the thresholded image of f(x,y).

Connected component analysis (CCA) is used to identify contiguous objects in a mask

There are at least 3 problems with this:

1) Single pixel spots are counted as objects

2) Edge nuclei are counted

3) Some nuclei are touching

- Use the options in the Analyze particles... to remove small particles and exclude edge objects.
- To fill small regions that are not included during throsholding, use Process -> Binary -> Fill Holes.  
- To split overlapping objects, use Process-> Binary -> Watershed.

**Image Segmentation with Labkit**

**Labkit** is a user-friendly Fiji plugin for the segmentation of microscopy image data. (https://imagej.net/plugins/labkit/)

It offers easy to use manual and automated image segmentation routines that can be rapidly applied to single- and multi-channel images as well as timelapse movies in 2D or 3D.

It incorporated BigData Viewer visulazation, helping with large volume datasets.

Macro- recordable.

Exploits the power of graphics processing units (GPUs) to gain additional performance and can even be used on high performance computing clusters (HPC) for distributed processing of big image data.

**Installation**

1) Install Fiji

2) Install Labkit’s update site:

    (i) Start Fiji
    (ii) From the menu, select Help > Update
    (iii) Click on Manage Update Sites, then select “Labkit” in the table
    (iv) Close, click on Apply changes and then Ok

3) Restart Fiji. Find Labkit in the menu at Plugins › Labkit.

**Points to remember**

- Labkit’s pixel classification algorithm only needs very little training data.

- Providing a large amount of training data often results in a poorly performing classification.

- To get a good result use the smallest available brush size, draw thin lines.

- Pick a few representative objects in the image and for those mark foreground and background pixels (or whatever classes you are using)
